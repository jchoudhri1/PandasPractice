```python
exam_data ={
    'name':['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score':[12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts':[1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify':['yes', 'no', 'yes', 'no','no','yes','yes','no','no','yes']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f','g','h','i','j']

df = pd.DataFrame(exam_data, index = labels)
df
```

```python
### select rows where number of attempts is less than 2 and score greater than 15
print(df[(df['attempts'] < 2) & (df['score'] > 15)])
```

```python
### Count the rows and columns
print("The number of columns is: " + str(len(df.columns)))
print("The number of rows is: " + str(len(df.index)))
```

```python
### select rows where score is missing
df1 = df[df.isna().any(axis=1)]
df1
```

```python
### Select rows where score is between 15 & 20
print(df[(df['score'] <= 20) & (df['score'] > 15)])
```
